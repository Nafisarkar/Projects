# For numeric columns (Age, Purchase Amount, Loyalty Points), replace missing values with the mean or median

```python
ageMean = df['Age'].mean()
print("AgeMean :" , ageMean)

df['Age'].fillna(int(ageMean), inplace=True)
df
```

```python
avgPurchesAmount = df['Purchase Amount'].mean()
print("Average purches amount: ", avgPurchesAmount)


df['Purchase Amount'].fillna(int(avgPurchesAmount), inplace=True)
df
```

```python
medianLoyality = df['Loyalty Points'].median()
print("Median Loyalty Points: ", medianLoyality)

df['Loyalty Points'].fillna(medianLoyality, inplace=True)
df
```

# For categorical columns (Gender, Payment Method, Membership Level), replace missing values with the mode (most frequent value).

```python
frequentGender = df['Gender'].mode()[0]
print(frequentGender)

df['Gender'].fillna(frequentGender, inplace=True)
df
```

```python
frequentPayMethod = df['Payment Method'].value_counts().idxmax()
print(frequentPayMethod)

df['Payment Method'].fillna(frequentPayMethod, inplace=True)
df
```

```python
frequentMemberShip = df['Membership Level'].value_counts().idxmax()
print(frequentMemberShip)

df['Membership Level'].replace(frequentMemberShip, inplace=True)
df
```

# For textual columns (Feedback), implement text imputation techniques.

```python
df['Feedback'].fillna(method='pad', inplace=True)
df
```
